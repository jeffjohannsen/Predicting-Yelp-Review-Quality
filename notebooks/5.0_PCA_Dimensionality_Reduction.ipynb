## PCA

### Scale Data

```python
start = time.perf_counter()
```

```python
standard_scaler = StandardScaler()
X_train_scaled = standard_scaler.fit_transform(X_train)
X_test_scaled = standard_scaler.transform(X_test)
```

```python
end = time.perf_counter()

print("\nTrain and Test Data Scaled")
print(f"Preprocessing took {(end-start):.2f} seconds.")
```

### Perform PCA

```python
start = time.perf_counter()
```

```python
num_components = 5
```

```python
pca = PCA(n_components=num_components)
```

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
end = time.perf_counter()

print(f"PCA Fit and Transform Complete")
print(f"PCA took {(end-start)/60:.2f} minutes.")
```

### Explained Variance

```python
explained_variance = pca.explained_variance_ratio_
print(f'{explained_variance.sum()*100:.1f}% of the variance is explained.')
```

### Scree Plot

PCA number of components must be "None".

```python
total_variance = np.sum(pca.explained_variance_)
cumulative_variance = np.cumsum(pca.explained_variance_)
proportion_variance_explained = cumulative_variance/total_variance
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(proportion_variance_explained, color = 'black', linewidth=2, label='Explained Variance')
ax.axhline(0.9, label='90% Goal', linestyle='--', linewidth=1)
ax.set_ylabel('Proportion of Explained Variance')
ax.set_xlabel('Number of Principal Components')
ax.legend();
```

## Create New Train and Test Dataframes

PCA number of components must not be "None".

```python
train_id_target = train[['review_id', 'target_clf', 'target_reg']]
train_principal_components = pd.DataFrame(X_train_pca, columns=[f'pc{i}' for i in range(1, num_components+1)])
train_pca = pd.concat([train_id_target, train_principal_components], axis=1) 
```

```python
test_id_target = test[['review_id', 'target_clf', 'target_reg']]
test_principal_components = pd.DataFrame(X_test_pca, columns=[f'pc{i}' for i in range(1, num_components+1)])
test_pca = pd.concat([test_id_target, test_principal_components], axis=1)
```
